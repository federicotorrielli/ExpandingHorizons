```python
column = 'GIOV P@10'
resource = 'CONCEPTNET-USED FOR'
filename = 'stats_conceptnet-UsedFor_at10.csv'

with open('../eval_stats/'+filename, 'r') as fr:
    i = fr.readline().split(',').index(column)
    print(column, '--> column', i+1)
    data = [float(line.strip().split(',')[i]) for line in fr if 'WIN ' not in line]
    fr.seek(0)
    fr.readline()
    names = [line.strip().split(',')[1].split('-')[0]+'-'+'_'.join(line.strip().split(',')[1].split('_')[-4:]) for line in fr if 'WIN ' not in line]
print(data)
print(names)
# Generate random data
# data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot a histogram
plt.hist(data, bins=30, alpha=0.5, color='blue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot a box plot
plt.boxplot(data, vert=True, patch_artist=True)  # vert=False for horizontal boxplot
plt.title('Box Plot of Normally Distributed Data')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # Seaborn is built on top of matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics

# Plot a density plot
sns.kdeplot(data, fill=True, color='blue')
plt.title('Density Plot of '+column)
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot a scatter plot
plt.figure(figsize=(18, 8), dpi=80)
plt.scatter(x=names, y=data)
plt.xticks(rotation=90)
plt.title('Scatter Plot showing '+column+' for '+resource)
plt.ylabel(column)
plt.grid(True)
plt.show()
```
